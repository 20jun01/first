```python
def SOLVE(quiz_data):
    import time
    import os
    print("出題モードです")
    print("どのモードにしますか？")
    print("A: ランダム出題モード")
    print("B: 順番出題モード")
    #print("C: ジャンル別出題モード")なかなか労力が必要なので後で
    aa=input()
    num_cor=0
    if aa=="A" or aa=="a":
        import random
        P=[]
        for i in range(quiz_data["num_prob"]):
            P.append(i)
        random.shuffle(P)
        print("何問出題しますか？")
        while(1):
            num=input()
            if is_int(num)==False:
                print("有効な数値を入力してください。")
                continue
            num=int(num)
            break
        for i in range(quiz_data["num_prob"]):
            if i == num:
                print("規定問題数に達しました。")
                break
            print("ジャンル["+quiz_data["prob_type"][P[i]]+"]"+"正答率:"+quiz_data["per_cor"][P[i]]+"%")
            print("問題：",end="")
            print(quiz_data["prob"][P[i]])
            ans=input()
            quiz_data["time_sol"][P[i]]+=1
            if ans in quiz_data["ans"][P[i]]:
                print("正解！！")
                quiz_data["time_cor"][P[i]]+=1
                num_cor+=1
                quiz_data["per_cor"][P[i]]=round(quiz_data["time_cor"][P[i]]/quiz_data["time_sol"][P[i]],2)*100
            else:
                print("不正解！")
                print("解答例："+quiz_data["ans"][P[i]][0])
                print("他には…")
                for m in range(1,len(quiz_data["ans"][P[i]])):
                    print(quiz_data["ans"][P[i]][m])
            #print("解説："+quiz_data[""][P[i]])
            print("続けますか？")
            print("はい:Y       いいえ:N")
            sel=input()
            if sel=="Y" or sel=="y" or sel=="YES" or sel=="Yes" or sel=="yes" or sel=="はい":
                print("クイズを続行します。")
                time.sleep(1)
                os.system("cls")
            elif sel=="N" or sel=="n" or sel=="No" or sel=="NO" or sel=="no" or sel=="いいえ":
                state=1
                break
            else:
                print("正常な入力ではなかったためクイズを続行します。")
        if i<=num and not state==1:
            print("問題が底をつきましたorz")
        print("クイズの出題を終了します。")
        percor=round(i/num_cor,2)*100
        s="結果"
        s=s.center(8,"―")
        print(s)
        print("解いた問題数："+i)
        print("正解した問題数:"+num_cor)
        print("正答率:"+per_cor)
        json.dump(quiz_data,f,indent=4)
    if aa=="B" or aa=="b":
        for i in range(quiz_data["num_prob"]):
            print("ジャンル["+quiz_data["prob_type"][i]+"]"+"正答率:"+quiz_data["per_cor"][i]+"%")
            print("問題：",end="")
            print(quiz_data["prob"][i])
            ans=input()
            quiz_data["time_sol"][i]+=1
            if ans in quiz_data["ans"][i]:
                print("正解！！")
                quiz_data["time_cor"][i]+=1
                num_cor+=1
                quiz_data["per_cor"][i]=round(quiz_data["time_cor"][i]/quiz_data["time_sol"][i],2)*100
            else:
                print("不正解！")
                print("解答例："+quiz_data["ans"][i][0])
                print("他には…")
                for m in range(1,len(quiz_data["ans"][i])):
                    print(quiz_data["ans"][i][m])
            #print("解説："+quiz_data[""][i])
            print("続けますか？")
            print("はい:Y       いいえ:N")
            sel=input()
            if sel=="Y" or sel=="y" or sel=="YES" or sel=="Yes" or sel=="yes" or sel=="はい":
                print("クイズを続行します。")
                time.sleep(1)
                os.system("cls")
            elif sel=="N" or sel=="n" or sel=="No" or sel=="NO" or sel=="no" or sel=="いいえ":
                state=1
                break
            else:
                print("正常な入力ではなかったためクイズを続行します。")
        if not state =1:
            print("全ての問題の出題が終了しました。")
        print("クイズの出題を終了します。")
        percor=round(i/num_cor,2)*100
        s="結果"
        s=s.center(8,"―")
        print(s)
        print("解いた問題数："+i)
        print("正解した問題数:"+num_cor)
        print("正答率:"+per_cor)
        json.dump(quiz_data,f,indent=4)
    else:
        print("選択肢から選んで下さい。")
        SOLVE(quiz_data)
```

```python
def APPEND(quiz_data):
    print("問題入力モードです。")
    print("sを入力で終了します。")
    while True:
        b=[]
        print("問題文を入力してください。")
        a=input()
        if a == "s":
            break
        print("問題の種類を入力してください。")
        t=input()
        if t=="s":
            break
        print("解答は何種類ありますか？")
        tem=input()
        if tem=="s":
            break
        num=int(tem)
        for i in range(num):
            print("解答を入力してください。")
            b.append(input())
            if b[i]=="s":
                break
        quiz_data["num_prob"]+=1
        quiz_data["prob"].append(a)
        quiz_data["prob_type"].append(t)
        quiz_data["ans"].append(b)
        quiz_data["num_ans"].append(num)
        quiz_data["per_cor"].append(0)
        quiz_data["time_cor"].append(0)
        quiz_data["time_sol"].append(0)
    json.dump(quiz_data,f,indent=4)
    print("終了しました。")
```
